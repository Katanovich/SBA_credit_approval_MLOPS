```python
def crop_image(img):

    #Determine height and width
    h = img.shape[0]
    w = img.shape[1]

    #Determine centre pixel
    Ch = h//2
    Cw = w//2

    #Determine the first non-black pixel directly above the center pixel at the top
    P = 0
    for i in range(Ch):
        if not(all(img[i][Cw] == 0)):
            P = i
            break;

    #Determine the radius of the retina
    R = Ch - P

    #Determine the boundaries for cropping
    X1 = (Ch-R)
    X2 = (Ch+R)
    Y1 = (Cw-R)
    Y2 = (Cw+R)

    img_cropped = []
    #Crop the image
    for i in range(3):
        img_cropped.append(img[X1:X2,Y1:Y2,i])

    img_cropped = np.dstack(img_cropped)

    return img_cropped
```

```python
def crop_image_from_gray(img, tol=7): #applies a mask to the image and returns a preprocessed images with 3 channels
    # For Grayscale images
    if img.ndim == 2:
        mask = img > tol
        return img[np.ix_(mask.any(axis = 1),mask.any(axis = 0))] # mask.any(axis = 1) makes a boolean array where each element in the array corresponds to each row in the image matrix. For a given row, the corresponding boolean value in the array is true if any value in the row is true.
                                                                  # mask.any(axis = 0) makes a boolean array where each element in the array corresponds to each col in the image matrix. For a given col, the corresponding boolean value in the array is true if any value in the col is true.
                                                                  # np.ix_(mask.any(axis = 1),mask.any(axis = 0)) gets those pixels from the image for which both the row and the column value is true.

    # If we have a normal RGB images
    elif img.ndim == 3:
        gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) #RGB image to grayscale
        mask = gray_img > tol #creates a boolean matrix

        check_shape = img[:,:,0][np.ix_(mask.any(1),mask.any(0))].shape[0]
        if (check_shape == 0): # If whole image is cropped because it was too dark
            return img # return original image
        else:
            img1=img[:,:,0][np.ix_(mask.any(axis = 1),mask.any(axis = 0))] #apply mask to pixel 0
            img2=img[:,:,1][np.ix_(mask.any(axis = 1),mask.any(axis = 0))] #apply mask to pixel 1
            img3=img[:,:,2][np.ix_(mask.any(axis = 1),mask.any(axis = 0))] #apply mask to pixel 2
            img = np.stack([img1,img2,img3],axis=-1)
        return img
```

```python
IMG_SIZE = 256
def preprocess(image_path):
    image = cv2.imread(image_path) #read image
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #convert BGR image to RGB
    image = crop_image_from_gray(image) #crop image
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE)) #resize image
    image = cv2.addWeighted (image, 4, cv2.GaussianBlur(image, (0,0) ,10), -4, 128) #add gaussian blur
    return image
```

```python
# Assuming x_train and x_val are Series of id_codes
# Retrieve original 1D diagnosis Series from df for one-hot encoding
original_y_train_series = df.set_index('id_code').loc[x_train]['diagnosis']
original_y_val_series = df.set_index('id_code').loc[x_val]['diagnosis']

# Preprocessing train images
train_images = np.empty((len(x_train),IMG_SIZE,IMG_SIZE,3), dtype='uint8') #empty numpy array
for i,img_id in enumerate(notebook.tqdm(x_train, desc="Preprocessing train images")):
    current_image_path = TRAIN_IMAGES / (img_id + ".png") # Construct path for current image
    train_images[i] = preprocess(str(current_image_path)) #preprocessing each image and storing it in the array

y_train_onehot = pd.get_dummies(original_y_train_series).values

# Preprocessing val images
val_images = np.empty((len(x_val), IMG_SIZE, IMG_SIZE, 3), dtype='uint8')
for i, img_id in enumerate(notebook.tqdm(x_val, desc="Preprocessing val images")):
    current_image_path = TRAIN_IMAGES / (img_id + ".png")
    val_images[i] = preprocess(str(current_image_path))

y_val_onehot = pd.get_dummies(original_y_val_series).values
```

```python
input1 = layers.Input((IMG_SIZE,IMG_SIZE,3))
x = layers.Conv2D(filters=32,kernel_size=3,activation='relu')(input1) #convolutional layer
x = layers.MaxPooling2D(pool_size=2)(x) #max pooling layer
x = layers.Dropout(0.2)(x) #dropout layer
x = layers.Flatten()(x) #flattening layer
x = layers.Dense(32,activation='relu')(x)
output1 = layers.Dense(5,activation = 'softmax')(x) #softmax layer

model1=tf.keras.Model(inputs=input1, outputs=output1) #initailizing the model

model1.compile(loss='categorical_crossentropy', #compiling the model
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=[
        'accuracy',
        tf.keras.metrics.AUC(multi_label=True, name='roc_auc')
    ]

              )

history = model1.fit(train_images, y_train_onehot, #training the model
                     steps_per_epoch= len(y_train_onehot) // BATCH_SIZE,
                     epochs=20,
                     validation_data = (val_images,y_val_onehot),
                     validation_steps = len(y_val_onehot) // BATCH_SIZE,
                     # callbacks = cb
                     )
```
